# Realized GARCH Model (Hansen et al., 2012)

The **Realized GARCH** model provides a flexible and rich representation for modeling returns and realized measures of volatility, as proposed by Hansen et al. (2012). This model explicitly relates the observed realized measure to the latent volatility using a measurement equation, while allowing for asymmetric reactions to shocks.

---

## **Model Specification**

### **1. Mean Equation**

The return process $y_t$ is modeled as:

$$
y_t = \mu_t + \sigma_t z_t, \quad z_t \sim i.i.d \, N(0, 1)
$$

Where:
- $y_t$: The asset return at time $t$,
- $\mu_t$: The conditional mean,
- $\sigma_t$: The conditional standard deviation,
- $z_t$: A standard normal innovation.

---

### **2. Volatility Equation**

The log of the conditional variance $\log \sigma_t^2$ evolves as follows:

$$
\log \sigma_t^2 = \omega + \sum_{i=1}^q \alpha_i \log r_{t-i} + \sum_{i=1}^p \beta_i \log \sigma_{t-i}^2
$$

Where:
- $\omega$: The intercept term,
- $\alpha_i$: Coefficients for the lagged realized measure $\log r_{t-i}$,
- $\beta_i$: Coefficients for the lagged log conditional variance $\log \sigma_{t-i}^2$,
- $q$ and $p$: Lag orders.

---

### **3. Measurement Equation**

The realized measure $\log r_t$ is related to the latent log variance $\log \sigma_t^2$ as:

$$
\log r_t = \xi + \delta \log \sigma_t^2 + \tau(z_t) + u_t, \quad u_t \sim N(0, \lambda)
$$

Where:
- $\xi$: The intercept term,
- $\delta$: The slope coefficient linking the realized measure to the latent volatility,
- $\tau(z_t)$: The asymmetric reaction function,
- $u_t$: Measurement error with variance $\lambda$.

---

### **4. Asymmetric Reaction Function**

The asymmetric reaction to shocks $\tau(z_t)$ is defined as a simple quadratic function:

$$
\tau(z_t) = \eta_1 z_t + \eta_2 (z_t^2 - 1)
$$

Where:
- $\eta_1$: Captures the linear response to shocks,
- $\eta_2$: Captures the quadratic response to deviations in shock magnitude.
